# now play with csv file

```
gender=pd.read_csv("GenderSequencesSwapped.csv")
```

```
gender.head()
```

```
gender["text_bert"]=''
gender["text_flipped_bert"]=''
gender["label"]=''
gender["flipped_label"]=''
gender["if_change"]=''
```

This is for masking the target word and predicting it again.
Prediction is saved in "label" column.

```
for i in range(0,len(gender)):
  gender.text_bert[i]=str(gender.sequence[i])+" "+str(gender.target[i])
  tokenized_text = tokenizer.tokenize(gender.text_bert[i]) 
  mask_idx = len(tokenizer.tokenize(gender.text_bert[i]))-1
  tokenized_text[mask_idx] = mask_tok
  tok_idxs = tokenizer.convert_tokens_to_ids(tokenized_text)
  seg_idxs = [0] * len(tok_idxs)
  tok_tensor = torch.tensor([tok_idxs])
  seg_tensor = torch.tensor([seg_idxs])
  preds = model(tok_tensor, seg_tensor)
  pred_idx = torch.argmax(preds[0, mask_idx]).item()
  pred_tok = tokenizer.convert_ids_to_tokens([pred_idx])[0]
  gender.label[i]=pred_tok
```

This is for masking the target word for the flipped sequences and predicting it again.
The prediction is saved in the "flipped_label" column.

```
for i in range(0,len(gender)):
  gender.text_flipped_bert[i]=str(gender.flipped[i])+" "+str(gender.target[i])
  tokenized_text = tokenizer.tokenize(gender.text_flipped_bert[i])
  mask_idx = len(tokenizer.tokenize(gender.text_flipped_bert[i]))-1
  tokenized_text[mask_idx] = mask_tok
  tok_idxs = tokenizer.convert_tokens_to_ids(tokenized_text)
  seg_idxs = [0] * len(tok_idxs)
  tok_tensor = torch.tensor([tok_idxs])
  seg_tensor = torch.tensor([seg_idxs])
  preds = model(tok_tensor, seg_tensor)
  pred_idx = torch.argmax(preds[0, mask_idx]).item()
  pred_tok = tokenizer.convert_ids_to_tokens([pred_idx])[0]
  gender.flipped_label[i]=pred_tok
```

Here we compare the gender of predictions.

```
for i in range(0,len(gender)):
  if gender.label[i] in DEFAULT_MALE_NOUNS and gender.flipped_label[i] in DEFAULT_FEMALE_NOUNS:
    gender.if_change[i]="1"
  elif gender.label[i] in DEFAULT_FEMALE_NOUNS and gender.flipped_label[i] in DEFAULT_MALE_NOUNS:
    gender.if_change[i]="1"
  else:
    gender.if_change[i]="0"
```
